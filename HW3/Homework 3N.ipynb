Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
import re
````

````python
string1 = "Call me at 382-384-3840." 
string2 = "my number is (510) 849-3519. Call me!"
string3 = "my number is 510-849-35192" 
string4 = "here’s my number: 510-849.3519"

# YOUR CODE HERE
# Dùng regex để tìm số điện thoại
pattern = r"(\(\d{3}\)\s|\d{3}-)\d{3}-\d{4}\b"

# Kiểm tra từng chuỗi
print(re.search(pattern, string1) is not None)  
print(re.search(pattern, string2) is not None)  
print(re.search(pattern, string3) is not None)  
print(re.search(pattern, string4) is not None)  
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
date_list = ["03/12/2018", "03.13.18", "03/14/2018", "03:15:2018"]

# YOUR CODE HERE
# regex để tìm các định dạng ngày
pattern = r'(\d{2})[\/.:\s](\d{2})[\/.:\s](\d{2,4})'

# Hàm thay thế để chuẩn hóa định dạng ngày
def replace_dates(match):
    day = match.group(1)
    month = match.group(2)
    year = match.group(3)
    if len(year) == 2:  # Nếu năm đề cho chỉ có hai chữ số
        year = '20' + year  # Giả sử đề cho năm trong thế kỷ 21
    return f"{day}-{month}-{year}"

# Duyệt qua từng ngày trong list đề cho và áp dụng re.sub
formatted_dates = [re.sub(pattern, replace_dates, date) for date in date_list]

# In ra kết quả theo dòng
for date in formatted_dates:
    print(date)
````

In this exercise, you you working with ```email_test.txt``` file (attached), using Regular Expression. 
1. Extract name and email address of sender  
Output should be something like this

2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004 


````python
fh = open(r"email_test.txt", mode="r", encoding="utf-8").read()
````

````python
# Extract name and email address of sender
# YOUR CODE HERE

# Trích xuất tên
names = re.findall(r'From: "?([\w\s\.]+)"?\s*<', fh)

# Trích xuất địa chỉ email
emails = re.findall(r'<([\w\.-]+@[\w\.-]+)>', fh)

# In kết quả
print("Name:")
for name in names:
    print(f'"{name}"')

print("\nEmail address:")
for email in emails:
    print(f'"{email}"')
````

````python
# Find SUBJECT of All emails from 01/06/2003 to 13/05/2004
import re
from datetime import datetime

# Đọc file txt
with open(r"email_test.txt", mode="r", encoding="utf-8") as file:
    fh = file.read()

# Tìm kiếm các cặp ngày và chủ đề bằng regex
pattern = r'Date: (\w{3}, \d{1,2} \w{3} \d{4}).*?\nSubject: (.+?)(?=\n\w|$)'
matches = re.findall(pattern, fh, re.DOTALL)

# Định nghĩa khoảng thời gian cần lọc
start_date = datetime.strptime('01/06/2003', '%d/%m/%Y')
end_date = datetime.strptime('13/05/2004', '%d/%m/%Y')

# Lọc các email trong khoảng thời gian đó
subjects = []
for date_str, subject in matches:
    try:
        date = datetime.strptime(date_str, '%a, %d %b %Y')  # Ngày theo định dạng: "Fri, 6 Jun 2003"
    except ValueError:
        continue  # Bỏ qua nếu không thể phân tích
    if start_date <= date <= end_date:
        subjects.append(subject.strip())

# In kết quả là các chủ đề trong khoảng thời gian cần lọc
for subject in subjects:
    print(f"['{subject}']")


````
