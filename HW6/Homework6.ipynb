```python
import pandas as pd
import numpy as np
```

```python
# Đọc Dataframe từ file csv, với các cột như sau
field_names = ['State', 'Sex', 'Year', 'Name', 'Count']
babynames_df = pd.read_csv('Data/STATE.CA.TXT',header=None,names=field_names)
```

```python
# Show 10 dòng ngẫu nhiên
print("10 dòng ngẫu nhiên:")
babynames_df.sample(10)
```

```python
# Hiển thị DF chỉ gồm female 
print("\nDataFrame chỉ gồm female:")
babynames_df[babynames_df['Sex'] == 'F']
```

```python
# Hiển thị DF gồm những người male, sinh trước 2000
print("\nDataFrame gồm những người male, sinh trước 2000:")
babynames_df[(babynames_df['Sex'] == 'M') & (babynames_df['Year'] < 2000)]
```

```python
# Hiển thị 3 babynames với count > 250
print("\n3 babynames với Count > 250:")
babynames_df[babynames_df['Count'] > 250].drop_duplicates(subset= 'Name').head(3)
```

```python
# Hiển thị các babynames có tên là Bella, Alex, Narges, hoặc Lisa
print("\nBabynames với tên Bella, Alex, Narges hoặc Lisa:")
babynames_df[babynames_df['Name'].isin(['Bella', 'Alex', 'Narges', 'Lisa'])]
```

```python
# Hiển thị với những Name bắt đầu bằng chữ N
print("\nBabynames bắt đầu bằng chữ N:")
babynames_df[babynames_df['Name'].str.startswith('N')]
```

```python
# Thêm cột name_lengths với giá trị là số lượng ký tự của cột Name
print("\nDataFrame sau khi thêm cột name_lengths:")
babynames_df['name_lengths'] = babynames_df['Name'].str.len()
babynames_df.head()
```

```python
# Hãy cho biết số lượng baby có tên 'Alice' trung bình hàng năm là bao nhiêu
alice_avg_count = babynames_df[babynames_df['Name'] == 'Alice']['Count'].mean()
print(f"\nSố lượng baby có tên 'Alice' trung bình hàng năm: {alice_avg_count}")
```

```python
# Hiển thị tất cả các tên trong DF, mỗi tên chỉ hiển thị 1 lần
print("\nTất cả các tên, mỗi tên chỉ hiển thị 1 lần:")
print(babynames_df['Name'].unique())
```

```python
# Tìm babynames phổ biến nhất trong năm 2020
most_popular_2020 = babynames_df[babynames_df['Year'] == 2020].sort_values('Count', ascending=False).iloc[0]
print(most_popular_2020)
print("\nBabynames phổ biến nhất trong năm 2020:", most_popular_2020.Name)
```
